```python
def moving_average(series, n):
    return np.average(series[-n:])
```

```python
moving_average(s_y, 25)
```

```python
def plot_moving_average(series, n):
    rolling_mean = series.rolling(window=n).mean()
    plt.figure(figsize=(15,5))
    plt.plot(rolling_mean, 'r')
    
    plt.plot(df[n:], 'b')
    plt.grid(True)
```

```python
plot_moving_average(df, 24)
```

```python
plot_moving_average(df, 24*7)
```

```python
def weighted_regression(series, weights):
    result = 0.0
    weights.reverse()
    for n in range(len(weights)):
        result += weights[n] * series[-n-1]
    return result
```

```python
weighted_regression(df['Users'], list(np.linspace(0.1, 1, num=24)[::-1]))
```

```python
weights = list(np.linspace(0.1, 1, num=24))
def plot_weighted(series, n):
    rolling_mean = weighted_regression(series.rolling(window=n, win_type='slepian').sum(), weights)
    
    plt.figure(figsize=(15,5))
    plt.plot(rolling_mean, 'r')
    
    plt.plot(df[n:], 'b')
    plt.grid(True)
```

```python
plot_weighted(df, 24)
```
